```python

# Para convertir de Ascii a binario primero se convierte el texto a un arreglo de caracteres donde cada caracteres es un valor númerico
text = "krnwenwrmxbjlroamxb"
alfabeto = 26
arreglo = list(text)

# Existe una función en python que puede convertir el texto a valor númerico en Ascii
for i in range( 0  , len(arreglo) ):
    arreglo[i] = ord(arreglo[i])

def decimal_a_binario(n):
    # si es cero es cero 
    if n == 0:
        return "0"
    binario = ""
    # https://www.geeksforgeeks.org/python/python-program-to-covert-decimal-to-binary-number/
    # Se saco de geek for greeks pero el tema es que se divide entre 2 hasta conseguir el residuo  
    while n > 0:
        binario = str(n % 2) + binario
        n //= 2
    return binario

string_binario = ""
for i in arreglo:
    binario = decimal_a_binario(i)
    string_binario = string_binario +" "+  binario


print("texto en binario ", string_binario)
```

Realizar la creación de un script que permita la conversión de BINARIO a BASE64

```python

binario = ['1100001', '110011', '1001010', '1110101', '1100100', '110010', '1010110', '1110101', '1100100', '110011', '1001010', '1110100', '1100101', '1000111', '1001010', '1110001', '1100010', '1001000', '1001010', '1110110', '1011001', '1010111', '110001', '110100', '1011001', '1100111', '111101', '111101']

decimal = []

# Function to convert 
# binary to decimal

    # https://www.geeksforgeeks.org/dsa/program-binary-decimal-conversion/
    # Se saco de geek for greeks pero el tema es que se divide entre 2 hasta conseguir el residuo
    # el se cuenta del último digito elevado al valor del espacio encontrado osea si se encontro el digito en la cuarta posición
    # se eleva 2^4  
def binaryToDecimal(n):
    num = n
    dec_value = 0
    
    # Initializing base 
    # value to 1, i.e 2 ^ 0
    base1 = 1
    
    len1 = len(num)
    for i in range(len1 - 1, -1, -1):
        if (num[i] == '1'):     
            dec_value += base1
        base1 = base1 * 2
    
    return dec_value


#Puedo juntar este for con la de abajo pero. 
for i in range(0, len(binario)):
    decimal.append( binaryToDecimal(binario[i]))

#Este for saca de todas las decimales a caracteres
for i in range(0, len(decimal)):
    decimal[i] = chr(binaryToDecimal(binario[i]))


#este for es utilizado para concatenar el text
texto = ""
for i in decimal:
    texto = texto + i

print(texto)

```
